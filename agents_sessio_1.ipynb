```python
#CREWAI llibreria per instanciar i definir els agents
from crewai import Agent, Task, Crew, LLM

#LLAMA 3 com a llenguatge LLM instal·lat en local
llm_local = LLM(
    model="ollama/llama3.2",
    base_url="http://localhost:11434" # Port per defecte de Llama3
)

#Exemple 1: Estudiant d'IA que cerca informació

#Inicialitzem l'agent
agentIA = Agent(
    role='Estudiante de Inteligencia Artificial',
    goal='Identificar y resumir los algoritmos de inteligencia artificial más utilitzados',
    backstory='Eres un estudiante de inteligencia artificial movido por la curiosidad de dominar los algoritmos más destacados y los modelos de última generación que actualmente dominan la disciplina',
    verbose=True,
    allow_delegation=False,
    llm=llm_local # Llama3.2
)

#Definim la tasca a realitzar
tasca1 = Task(
    description='Realizar una revisión de los algoritmos o modelos de intel·ligencia artificial de más impacto y uso actual',
    expected_output='Para elaborar la respuesta, selecciona los 3 principales indicando para cada uno de ellos: 1) Breve resumen de su funcionamiento, 2) Sus aplicaciones en el mundo real, y 3) El motivo por el que se consideran de última generación',
    agent=agentIA
)

#Inicialitzem l'equip
equip = Crew(
    agents=[agentIA],
    tasks=[tasca1],
    verbose=True
)

#Executem
resultat = equip.kickoff()
print(resultat)
```

```python
#Exemple 2: Equip d'anàlisi de dades
#Dataset: https://www.kaggle.com/datasets/suraj520/customer-support-ticket-dataset
import pandas as pd

dataset = pd.read_csv('customer_support_tickets.csv')

#inspeccionem les dades
dataset.head(3)
```

```python
from crewai.tools import tool

#Definim com a eina el propi dataset
@tool("Analizador del Dataset de Tiquets")
def analyze_tickets_csv(file_path = 'customer_support_tickets.csv') -> str:
    """
    Llegeix l'arxiu CSV de tiquets de suport i calcula els KPIs principals
    """
    try:
        df = pd.read_csv(file_path)     
        total_tickets = len(df)
        high_priority = len(df[df['Priority'] == 'High'])
        
        # Calcular el temps mitjà de resolució per departament
        avg_res_time = df.groupby('Department')['Resolution_Time_Hours'].mean().to_dict()
        
        # Calcular el volum de tiquets per departament
        ticket_volume = df['Department'].value_counts().to_dict()

        # Construir l'informe de text que llegirà l'Agent
        report = f"RESUM DE DADES EXTRETES DE {file_path}:\n"
        report += f"- Total de tiquets gestionats: {total_tickets}\n"
        report += f"- Tiquets de prioritat ALTA: {high_priority}\n\n"
        report += "Mètriques per Departament:\n"
        
        for dept in avg_res_time.keys():
            report += f"  * Departament: {dept} | Volum: {ticket_volume.get(dept, 0)} tiquets | Temps mitjà de resolució: {avg_res_time[dept]:.2f} hores\n"
            
        return report
    except Exception as e:
        return f"Error processant l'arxiu CSV. Detalls: {e}"
```

```python
#Definim els agents implicats
# Agent 1: Analista de Dades
data_analyst = Agent(
    role='Analista de Datos Sénior',
    goal='Extraer y resumir los KPIs críticos del dataset de tiquets de soporte',
    backstory='Eres un experto en limpiar datos y calcular métricas. Tu trabajo es leer los datos en bruto y trasladarlas a números claros para el resto del equipo',
    verbose=True,
    allow_delegation=False,
    tools=[analyze_tickets_csv],
    llm=llm_local 
)

# Agent 2: Optimitzador de processos
process_analyst = Agent(
    role='Analista de Operaciones y Procesos',
    goal='Identificar los cuellos de botella e ineficiencias basándonos en los KPIs',
    backstory='Tienes una mente analítica que te permite detectar dónde fallan los procesos de una empresa. Recibes métricas y eres capaz de deducir por qué un departamento tarda demasiado o está colapasado, y procesas soluciones prácticas y eficientes',
    verbose=True,
    allow_delegation=False,
    llm=llm_local
)

# Agent 3: Creador de l'informe de direcció
ceo_reporter = Agent(
    role='Especialista en Reporting Ejecutivo',
    goal='Redactar un informe final visual, directo y enfocado a negocio para el CEO',
    backstory='Eres el nexo entre los datos técnicos y la dirección. Recoges las conclusiones del resto del equipo y las transformas en un documento de diseño atractivo y fácil lectura, destacando los problemas más graves y las acciones a emprender',
    verbose=True,
    allow_delegation=False,
    llm=llm_local
)
```

```python
# Tasca 1: anàlisi de dades
task_data_extraction = Task(
    description='Utiliza la herramienta disponible para analizar el dataset de customer support ticket proporcionado. Extrae los KPIs de volumen y tiempo de resolución por departamento',
    expected_output='Un listado claro de KPIs con el volument de tiquets y los tiempos de resolución medio por departamento',
    agent=data_analyst
)

# Tasca 2: anàlisi de processos
task_process_analysis = Task(
    description='Revisa los KPIs proporcionadospor el Analista de Datos. Busca desviaciones relevantes (por ejemplo, un departamento que tarda mucho más que el resto). Identifica cuál es el cuello de botella y escribe 3 propuestas de mejora viables',
    expected_output='Un análisis de los cuellos de botella y una lista de 3 recomendaciones operativas',
    agent=process_analyst
)

# Tasca 3: elaboració de l'informe
task_executive_report = Task(
    description='Recoge las recomendaciones y los cuellos de botella detectados por el Analista de Procesos. Redacta un informe para el CEO en formato html. Utiliza tablas, gráficos, negritas y una estructura clara (Resumen Ejecutivo, Situación Actual, Cuellos de Botella y Plan de Acción)',
    expected_output='Un documento web (html) atractivo, de fácil lectura y listo para presentar a dirección',
    agent=ceo_reporter,
    output_file='informe_ceo.html' # Guardarà el resultat final automàticament
)
```

```python
#Definició del procés
from crewai import Process

support_crew = Crew(
    agents=[data_analyst, process_analyst, ceo_reporter],
    tasks=[task_data_extraction, task_process_analysis, task_executive_report],
    process=Process.sequential, # L'output d'un és l'input del següent
    verbose=True
)

# Executar el procés
print("Iniciant l'anàlisi de l'equip CrewAI...")
resultat_final = support_crew.kickoff()

print("PROCÉS FINALITZAT. L'informe s'ha guardat a 'informe_ceo.html'")
```
